```python
from bs4 import BeautifulSoup
import urllib2
wiki = "https://en.wikipedia.org/wiki/List_of_prisons_in_the_United_Kingdom"
header = {'User-Agent': 'Mozilla/5.0'} #Needed to prevent 403 error on Wikipedia
req = urllib2.Request(wiki,headers=header)
page = urllib2.urlopen(req)
soup = BeautifulSoup(page)
 

table = soup.find("table", { "class" : "wikitable sortable" })
```

```python
import pandas as pd
```

```python
df = pd.read_html(table.prettify(),header=0)[0]
```

```python
df.columns = ["wiki_name", "wiki_location", "wiki_county", "wiki_operator", "wiki_capacity", "wiki_notes", "wiki_category"]
```

```python
df.to_csv("wiki_prison_table.csv", index=False)
```
