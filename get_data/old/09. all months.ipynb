```python
import pandas as pd
```

```python
df_all_months_rtf = pd.read_csv("pop_data/all_pop_data_rtf.csv").sort(["date", "prison_name"])
df_all_months_txt = pd.read_csv("pop_data/all_pop_data_txt.csv").sort(["date", "prison_name"])
```

```python
df_all_months = pd.concat([df_all_months_rtf, df_all_months_txt])
```

```python
pd.options.display.max_columns = 10000
df_all_prisons = pd.read_csv("all_prison_data_manual.csv")
df_all_prisons.head()
```

```python
df_prison_details = df_all_prisons[["moj_prison_name", 'dt_lat', "dt_lng", "dt_type_of_prison", "dt_youth", "dt_category", "dt_other", "dt_address", "dt_region", "dt_full_name","gscraped_desc"]].fillna("")
for c in ["dt_category", "dt_youth", "dt_other"]:
    df_prison_details[c] = df_prison_details[c].str.strip()
df_prison_details.head()
```

```python
df_final = df_all_months.merge(df_prison_details, left_on="prison_name", right_on="moj_prison_name", how="left")
```

```python
df_final = df_final.rename(columns={"prison_name":"name", "dt_lat":"lat", "dt_lng":"lng"})
```

```python
df_final.to_csv("plot_data_all_months.csv", index=False, encoding="utf-8")
df_final.to_csv("../web/data/plot_data_all_months.csv", index=False)
```
