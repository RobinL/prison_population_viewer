```python
from get_one_month import get_one_month_from_rtf
url = "https://www.gov.uk/government/uploads/system/uploads/attachment_data/file/523261/monthly-bulletin-16-04.rtf"
month_df  = get_one_month_from_rtf(url)
month_df.head()
```

```python
import pandas as pd 
prison_df = pd.read_csv("all_prison_data_manual.csv")
prison_df.head()
```

```python
df_all = month_df.merge(prison_df, left_on="prison_name", right_on="moj_prison_name", how="left")
```

```python
df_all = df_all.rename(columns={"dt_lat":"lat", "dt_lng":"lng"})
```

```python
df_all.head()
```

```python
df_all.to_csv("plot_data.csv", index=False)
df_all.to_csv("../web/data/plot_data.csv", index=False)
```
