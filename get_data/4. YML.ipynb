```python
import yaml
import os
```

```python
rows = []
for i in os.listdir("prisons_yml"):
    if i.endswith(".yml"): 
        with open("prisons_yml/"+i) as f:
            y = yaml.load(f)
            try:
                y["address"] = ", ".join(y["address"])
            except:
                pass
            rows.append(y)
```

```python
import pandas as pd 

pd.DataFrame(rows)[["estate", "name", "nomis_id"]].to_csv("nomis_id_from_yaml.csv", index=False, encoding="utf-8")
```
